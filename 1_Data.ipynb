Numpy 


```python
import numpy as np
```

```python
# array

array1 = np.array([2, 3, 4])


# array in range between 0 and 20 with step of 2
array2 = np.arange(0,21,2)

print(array2)
```

```python
# zeroes
array3 = np.zeros((4,4))

print(array3)
```

```python
# ones
array4 = np.ones((4,4))
print(array4)
```

```python
array4_shape = array4.shape
print(array4_shape)
```

```python
array5 = array4.reshape((2, 8))

print(array5)
```

```python
# random

rand1 = np.random.rand(4)
print(rand1)

rand2 = np.random.randint(1,100)
print(rand2)
```

```python
# linspace (evenly spaced)
array6 = np.linspace(0,100,4)
print(array6)
```

```python
#Identity Matrix
array7 = np.eye(4)
print(array7)
```

```python
# selection

# (row, column)
print(array7[1])

print(array7[1, 1])

print(array7[2:, 2])
```

```python
# Math
print(array1 - array1)

print(array1 + array1)

print(array1 * array1)

print(array1 / array1)

print(np.sqrt(array1))

print(np.max(array1))

print(np.tan(array1))
```
